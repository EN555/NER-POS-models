```python
import json
with open("../NER-POS-models/score_by_filter_and_triple.json", "r") as file:
    score_by_filter_and_triple = json.load(file)
```

```python
noise_triple = ["t1 "," en", "end", "nd1", "nd2", " st", "sta", "tar", "art", "rt1", "rt2", "t2 ", "d2 ", "d1 ", '   ', "2  ", "1  "]
for filter_key in score_by_filter_and_triple.keys():
    scores = {}
    print(f"######## {filter_key} #####")
    for key,val in score_by_filter_and_triple[filter_key].items():
        if key not in noise_triple:
            scores[key] = sum(val)/len(val)
    print(sorted(scores, key=scores.get, reverse=True)[:10])
    print("\n\n\n")
```

```python
with open("../NER-POS-models/argmax_by_label_and_triple.json", "r") as file:
    argmax_by_label_and_triple = json.load(file)
noise_triple = [" en", "end", "nd1", "nd2", " st", "sta", "tar", "art", "rt1", "rt2", "t2 ", "d2 ", "d1 ", '   ', "2  ", "1  "]
for label, argmax_by_triple in argmax_by_label_and_triple.items():
    print(label)
    for key in noise_triple:
        argmax_by_triple.pop(key, None)
    tirples = sorted(argmax_by_triple, key=argmax_by_triple.get, reverse=True)[20:50]
    print(tirples)
```

```python
with open("../NER-POS-models/argmax_by_filter_label_triple.json", "r") as file:
    argmax_by_filter_label_triple = json.load(file)
noise_triple = [" en", "end", "nd1", "nd2", " st", "sta", "tar", "art", "rt1", "rt2", "t2 ", "d2 ", "d1 ", '   ', "2  ", "1  "]
```

```python
from collections import defaultdict
swapped_dict = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
for filter_index, label_dict in argmax_by_filter_label_triple.items():
    for label_name in label_dict.keys():
        swapped_dict[label_name][filter_index] = label_dict[label_name]
```

```python
for label, filter_triple in swapped_dict.items():
    print(f"#######{label}######")
    for filter_idx, triplets in filter_triple.items():
        print(f"------{filter_idx}------")
        print(sorted(triplets, key=triplets.get, reverse=True)[20:50])
```
